```python
import pandas as pd
```

```python
position = pd.read_csv("OpenPosition_12_13_2017.csv", index_col =0)
```

```python
position.head()
```

```python
long_position = position[position.Quantity>0]
```

```python
long_position.head()
```

```python
long_position['Weight'] = long_position["MarketValue"]/long_position.MarketValue.sum()
```

```python
long_position.head()
```

```python
long_position.sort("Weight", ascending = False).head()
```

```python
short_position = position[position.Quantity<0]
```

```python
short_position.head()
```

```python
short_position['Weight'] = -short_position["MarketValue"]/short_position.MarketValue.sum()
```

```python
short_position.head()
```
